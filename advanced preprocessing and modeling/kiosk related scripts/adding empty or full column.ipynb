```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv("./data/station_history_cumulative_weather.csv", index_col=0)
```

```python
data.head()
```

```python
def empty_full_or_neither(row):
    if row['bikes'] == 0:
        return 0
    elif row['docks'] == 0:
        return 1
    else:
        return 2


data['empty_full_or_neither'] = data.apply(empty_full_or_neither, axis = 1)
```

```python
data[data['empty_full_or_neither'] == 1]
```

```python
data.to_csv('./data/station_history_cumulative_weather.csv', index=False)
```
