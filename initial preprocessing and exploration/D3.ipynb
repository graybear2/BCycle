```python
import numpy as np
import pandas as pd
```

```python
trips_df = pd.read_csv('../metadata/Austin_B-Cycle_Trips_extended_ordered.csv', index_col=0)
trips_df.head()
```

```python
# GET ALL STATIONS NAMES
station_names = list()
for index, row in trips_df.iterrows():
    if row['Checkout Kiosk'] not in station_names:
        station_names.append(row['Checkout Kiosk'])
    if row['Return Kiosk'] not in station_names:
        station_names.append(row['Return Kiosk'])
print('Found {} unique stations'.format(len(station_names)))
```

```python
# CREATE EMPTY TRIP MATRIX
trip_matrix_df = pd.DataFrame(np.zeros((len(station_names), len(station_names))), 
        index=station_names, columns=station_names, dtype=np.int32)
trip_matrix_df.head()
```

```python
# POPULATE AND EXPORT TRIP MATRIX
for index, row in trips_df.iterrows():
    start = row['Checkout Kiosk']
    end = row['Return Kiosk']
    trip_matrix_df.at[start, end] += 1
trip_matrix_df.to_csv('../metadata/Trip_Matrix.csv')
```

```python
trip_matrix_df.head()
```

```python
numStations = 20
topTrips = list()
for i in range(numStations*2):
    topTrips.append(('', '', 0))

def insert(trips, entry):
    last = (numStations*2)-1
    if entry[2] > trips[last][2]:
        trips[last] = entry
    index = last-1
    while index >=0 and entry[2] > trips[index][2]:
        trips[index+1] = trips[index]
        trips[index] = entry
        index -= 1


for i in range(len(station_names)):
    for j in range(i, len(station_names)):
        check, ret = (station_names[i], station_names[j])
        insert(topTrips, (check, ret, trip_matrix_df.at[check, ret]))
        
topStations = list()
for trip in topTrips:
    if trip[0] not in topStations:
        topStations.append(trip[0])
    if len(topStations) >= numStations:
        break
    if trip[1] not in topStations:
        topStations.append(trip[1])
    if len(topStations) >= numStations:
        break

topStations_df = pd.DataFrame(np.zeros((numStations, numStations)), 
        index = topStations, columns = topStations, dtype=np.int32)
print("Found {} out of {} desired stations".format(len(topStations), numStations))
```

```python
# POPULATE AND EXPORT TOP TRIP MATRIX
for index, row in trips_df.iterrows():
    start = row['Checkout Kiosk']
    end = row['Return Kiosk']
    if start in topStations and end in topStations:
        topStations_df.at[start, end] += 1
topStations_df.to_csv('../metadata/Trip_Matrix_Top20.csv')
```

```python
total = 0
for index, row in topStations_df.iterrows():
    for entry in row:
        total += entry
print("{} out of {} total trips represented, {}%".format(total, trips_df.shape[0], 
        float(total)*100/float(trips_df.shape[0])))
```
